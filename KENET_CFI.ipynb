```python
# Normalizing the data
df = data.select_dtypes(include=['float64', 'int64'])

scaler = MinMaxScaler()

data_norm = scaler.fit_transform(df)

data_norm = pd.DataFrame(data_norm, columns = df.columns)

data_norm.sample()
```

```python
# Define the reference vector as 1 for all indicators
n_indicators = 111
X_star = np.ones(n_indicators)
```

```python
# Define the fuzzification constant of k
k = 1
```

```python
# Calculate the fuzzy metric of each indicator
M = np.zeros((47,111))

for j in range (111):
    M[:, j] = k * X_star[j] / (k * X_star[j] + abs(data_norm.iloc[:, j] - X_star[j]))
```

```python
# Calculate the composite fuzzy indicator for each observation 
CFI = np.prod(M, axis=1) ** (1/111)
```

```python
# Create dataframe with the CFI
CFI = pd.DataFrame(CFI, columns= ['CFI']) 

CFI_data = pd.concat([data.select_dtypes(include= 'object'), CFI], axis=1)

CFI_data['CFI Rank'] =  CFI_data['CFI'].rank(ascending=False)

CFI_data = CFI_data.sort_values('CFI', ascending=False)

CFI_data
```

```python
# Composite Indicator using PCA
gsa = pd.read_csv("Merged GSA.csv")
gsr = pd.read_csv("Merged GSR.csv")
```

```python
gsa = gsa.drop(columns=['Unnamed: 0', 'County'] ,axis=1)
```

```python
gsa.shape
```

```python
gsr = gsr.drop(columns=['Unnamed: 0', 'County'], axis=1)
```

```python
gsr.shape
```

```python
# Gsr Score
gsr_norm = scaler.fit_transform(gsr) # Normalize data

pca = PCA(1)

gsr_pca = pca.fit_transform(gsr_norm)

gsr_score = np.dot(gsr_norm, pca.components_.transpose())

gsr_score = pd.DataFrame(gsr_score, columns = ['GSR Score'])

gsr_score.head()

```

```python
#GSA Score
gsa_norm = scaler.fit_transform(gsa) # Normalize data

pca = PCA(1)

gsa_pca = pca.fit_transform(gsa_norm)

gsa_score = np.dot(gsa_norm, pca.components_.transpose())

gsa_score = pd.DataFrame(gsa_score, columns = ['GSA Score'])

gsa_score.head()
```

```python
# Get the weights of the each indicator
gsr_weight = len(gsr.columns)/len(df.columns)
gsa_weight = len(gsa.columns)/len(df.columns)

print("GSR Weight:", gsr_weight)
print("GSA Weight:", gsa_weight)
```

```python
# Generate the Index
combination = pd.concat([gsr_score, gsa_score], axis = 1)

for i in combination:
    combination["Index"] = ((gsr_weight * combination['GSR Score']) + gsa_weight * combination['GSA Score'])

combination.head()
```

```python
PCA_Index = pd.concat([data.select_dtypes(include= 'object'), combination['Index']], axis=1)

PCA_Index['PCA Rank'] =  PCA_Index['Index'].rank(ascending=False)

PCA_Index = PCA_Index.sort_values('Index', ascending=False)

PCA_Index
```

```python
PCA_Index.to_csv("PCA Index.csv")
CFI_data.to_csv("CFI.csv")
```

```python
# Arithmetic average rank
average = data_norm.mean(axis = 1)

Avg_Index = pd.concat([data.select_dtypes(include= 'object'), average], axis=1)

Avg_Index = Avg_Index.rename(columns={0: 'Average'})

Avg_Index['Average Rank'] = Avg_Index['Average'].rank(ascending=False)

Avg_Index = Avg_Index.sort_values('Average', ascending=False)

Avg_Index
```

```python
import numpy as np
data2 = data_norm + 1e-6

geometric_averages = np.prod(data2, axis=1) ** (1 / len(data2.columns))

geometric_averages = pd.concat([data.select_dtypes(include= 'object'), geometric_averages], axis=1)

geometric_averages = geometric_averages.rename(columns={0: 'Geo Average'})

geometric_averages['Geo Average Rank'] = geometric_averages['Geo Average'].rank(ascending=False)

geometric_averages = geometric_averages.sort_values('Geo Average', ascending=False)

geometric_averages
```

```python
# Comparing the results
merged_results = pd.merge(CFI_data, PCA_Index, on='County')

merged_results = pd.merge(merged_results, Avg_Index, on='County')

merged_results = pd.merge(merged_results, geometric_averages)

merged_results
```
